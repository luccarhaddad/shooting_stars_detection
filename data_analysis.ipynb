## 1. Seleção de features relevantes

Foram fornecidos 21 conjuntos de características para análise de dados, dos quais escolhemos os 5 a seguir:


- **AutoColorCorrelogram**: Captura a correlação espacial de cores nas imagens, sendo robusta a mudanças de aparência e forma. Isso é útil para distinguir meteoros com base na variação de cores em diferentes áreas da imagem, o que é crucial em um cenário dinâmico como o céu noturno.

- **CEDD**: Combina informações de cor e textura em um histograma compacto, permitindo uma representação rica de imagens astronômicas. A combinação dessas características é importante para capturar tanto a aparência visual quanto as texturas dos meteoros em contraste com o fundo.

- **FCTH**: Integra cor e textura em um histograma fuzzy, sendo eficaz mesmo em condições de distorção e ruído. Isso é relevante para imagens astronômicas, nas quais os meteoros podem aparecer com variações de forma e iluminação, exigindo uma característica robusta.

- **Haralick**: Usa matrizes de coocorrência para medir texturas, sendo ideal para capturar a variação de textura entre objetos brilhantes (meteoros) e fundos uniformes (céu). A textura é um fator-chave para diferenciar meteoros de ruído visual.

- **Gabor**: Analisa a frequência e a textura em direções específicas, semelhante ao sistema visual humano. Isso ajuda a identificar meteoros por suas características de movimento e variações de frequência em uma área específica da imagem.

```python
def removeColumnsOfZerosOrNAN(data, ignore_class=True):
    if ignore_class:
        ignored_columns = ['Class']
    else:
        ignored_columns = []
        
    data = data.drop('Unnamed: 0', axis=1)
    data_ignored = data[ignored_columns]
    data_to_process = data.drop(columns=ignored_columns, axis=1)
    columns_to_drop = data_to_process.columns[(data_to_process == 0).all()]
    data_to_process.drop(columns=columns_to_drop, inplace=True)
    final_data = pd.concat([data_to_process, data_ignored], axis=1)
    final_data.dropna(axis=1)
    return final_data
```

```python
def removeDuplicatedColumns(data, ignore_class=True):
    if ignore_class:
        ignored_columns = ['Class']
    else:
        ignored_columns = []
    
    data_ignored = data[ignored_columns]
    data_to_process = data.drop(columns=ignored_columns, axis=1)
    columns_to_drop = []
    
    for i in range(data_to_process.shape[1]):
        for j in range(i + 1, data_to_process.shape[1]):
            if data_to_process.iloc[:, i].equals(data_to_process.iloc[:, j]):
                columns_to_drop.append(data_to_process.columns[j])
    
    data_to_process.drop(columns=columns_to_drop, inplace=True)
    final_data = pd.concat([data_to_process, data_ignored], axis=1)
    return final_data
```

```python
def removeFullyCorrelated(data, threshold=0.95, ignore_class=True):
    if ignore_class:
        ignored_columns = ['Class']
    else:
        ignored_columns = []
    
    data_ignored = data[ignored_columns]
    data_to_process = data.drop(columns=ignored_columns, axis=1)
    corr_matrix = data_to_process.corr().abs()
    
    upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > threshold)]
    data_to_process.drop(columns=to_drop, inplace=True)
    
    final_data = pd.concat([data_to_process, data_ignored], axis=1)
    return final_data
```

```python
def minMaxScaling(data, ignore_class=True):
    if ignore_class:
        ignored_columns = ['Class']
    else:
        ignored_columns = []
    data_ignored = data[ignored_columns]
    data_to_process = data.drop(columns=ignored_columns, axis=1)
    data_normalized = (data_to_process - data_to_process.min()) / (data_to_process.max() - data_to_process.min())
    final_data = pd.concat([data_ignored, data_normalized], axis=1)
    final_data.dropna(axis=1)
    return final_data
```

```python
def dataPreProcessingPipeline(data, ignore_class=True):
    data = removeColumnsOfZerosOrNAN(data, ignore_class=ignore_class)
    data = removeDuplicatedColumns(data, ignore_class=ignore_class)
    data = removeFullyCorrelated(data, ignore_class=ignore_class)
    data = minMaxScaling(data, ignore_class=ignore_class)
    return data
```

## 2. Treinamento de classificadores preditivos 

```python
from sklearn.model_selection import cross_val_score, GridSearchCV, StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import log_loss, make_scorer
```

```python
X_train_AutoColorCorrelogram = train_AutoColorCorrelogram.drop(columns=['Class'])
y_train_AutoColorCorrelogram = train_AutoColorCorrelogram['Class']

X_train_CEDD = train_CEDD.drop(columns=['Class'])
y_train_CEDD = train_CEDD['Class']

X_train_FCTH = train_FCTH.drop(columns=['Class']).dropna(axis=1)
y_train_FCTH = train_FCTH['Class']

X_train_Haralick = train_Haralick.drop(columns=['Class'])
y_train_Haralick = train_Haralick['Class']

X_train_Gabor = train_Gabor.drop(columns=['Class'])
y_train_Gabor = train_Gabor['Class']
```

```python
datasets_train = [
    (X_train_AutoColorCorrelogram, y_train_AutoColorCorrelogram),
    (X_train_CEDD, y_train_CEDD),
    (X_train_FCTH, y_train_FCTH),
    (X_train_Haralick, y_train_Haralick),
    (X_train_Gabor, y_train_Gabor)
]

datasets_names = ['AutoColorCorrelogram', 'CEDD', 'FCTH', 'Haralick', 'Gabor']
```

```python
def evaluate_classifiers(datasets, datasets_names):
    results = []
    external_cv = StratifiedKFold(n_splits=5)

    for i, (X_train, y_train) in enumerate(datasets):
        dataset_name = datasets_names[i]
        
        #KNN
        knn = KNeighborsClassifier()
        param_grid = {'n_neighbors': [1, 3, 5]}
        internal_cv = StratifiedKFold(n_splits=3)
        grid_search_knn = GridSearchCV(knn, param_grid, cv=internal_cv, scoring=make_scorer(log_loss, greater_is_better=False))
        log_loss_knn = cross_val_score(grid_search_knn, X_train, y_train, cv=external_cv, scoring=make_scorer(log_loss, greater_is_better=False))
        
        # Árvores de Decisão
        dt = DecisionTreeClassifier()
        log_loss_dt = cross_val_score(dt, X_train, y_train, cv=external_cv, scoring=make_scorer(log_loss, greater_is_better=False))
        
        # Armazenar resultados formatados (média ± desvio padrão)
        knn_result = f"{np.mean(log_loss_knn):.2f} ± {np.std(log_loss_knn):.2f}"
        dt_result = f"{np.mean(log_loss_dt):.2f} ± {np.std(log_loss_dt):.2f}"

        results.append({
            'Conjunto': dataset_name,
            'kNN': knn_result,
            'AD': dt_result
        })
    return pd.DataFrame(results)
```

```python
results_df = evaluate_classifiers(datasets_train, datasets_names)
print(results_df)
```
